# Image Web Scrapping

```python
link = 'https://photos-stock.netlify.app/'
```

```python
service = Service(chromedriver_binary.chromedriver_filename)

diver = webdriver.Chrome(service = service)

diver.get(link)
```

```python
soup = BeautifulSoup(diver.page_source, 'html.parser')
```

```python
sp = soup.find('article', class_='photo')
```

```python
sp
```

```python
sp = soup.find('img').get('src')
```

```python
sp
```

```python
soup.find('h4').text
```

```python
soup.find('a').get('href')
```

```python
data=[]


for sp in soup.find_all('article'):
    info = sp.find('img').get('src')
    name = sp.find('h4').text
    likes = sp.find('p').text
    data.append([info,name,likes])
```

```python
data
```

```python
df = pd.DataFrame(data, columns =['Link','Name','Likes'])
```

```python
df.head()
```

```python
df.to_csv('Image.csv',index = False)
```
