# Scaping Quotes

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(html, 'html.parser')
```

```python
print(soup.find('span', class_ = 'text').text[1:-1])
```

```python
quotes=[]
for q in soup.find_all('span',class_='text'):
    quotes.append(q.text[1:-1])
                       
```

```python
quotes
```

```python
# Scrapping all quotes withs Quotes,Author names, about links and Tags

data=[]

for d in soup.find_all('div',class_='quote'):
    quote = d.find('span',class_='text').text[1:-1]
    author = d.find('small',class_='author').text
    about = d.find('a').get('href')
    tags=[]
    for t in d.find_all('a',class_='tag'):
        tags.append(t.text)
    data.append([quote,author,about,tags])
```

```python
data
```

```python
import pandas as pd
```

```python
# Stroing Data Into Data Frames

df = pd.DataFrame(data,columns=['Qoutes','Author','About Link','Tags'])

df.to_csv('Quotes.csv',index=False)
```

```python
from tqdm import tqdm

data=[]

for page in tqdm(range(1,6)):
    link = 'https://quotes.toscrape.com/page/'+str(page)
    res = requests.get(link)
    soup = BeautifulSoup(res.text, 'html.parser')
    for d in soup.find_all('div',class_='quote'):
        quote = d.find('span',class_='text').text[1:-1]
        author = d.find('small',class_='author').text
        about = d.find('a').get('href')
        tags=[]
        for t in d.find_all('a',class_='tag'):
            tags.append(t.text)
        data.append([quote,author,about,tags])
    
```
